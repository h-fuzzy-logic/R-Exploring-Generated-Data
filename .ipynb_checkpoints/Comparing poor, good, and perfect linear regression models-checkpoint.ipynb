### Introduction
The goal of this notebook is to use generated data to explore the accuracy of linear regression models.  In other words, when you have the advantage of knowing the "truth" because you generate it, trying different models to uncover the "truth" provides opportunities to dive into model summaries, diagnostics, and error calculations.  You can start to understand when a linear regression works and when it does not.  Based on:
* _Doing Data Science: Straight Talk from Frontline_ by Cathy O'Neil and Rachael Schutt.
* https://data.library.virginia.edu/diagnostic-plots/
* https://data.princeton.edu/r/linearmodels 
* http://www.sthda.com/english/articles/39-regression-model-diagnostics/161-linear-regression-assumptions-and-diagnostics-in-r-essentials/

### Steps
1. Fill a dataframe with generated data (to simulate a dataset in the wild)
2. For each truth, coefficients, diagnostics, and mean squared error for poor, good, and perfect models
    * truth1 is: y = beta0 + beta1*normal_distribution + error  
    * truth2 is: beta0 + beta1*gamma_distribution + error
    * truth3 is: normal_distribution^2

### Description of Diagnostics
|Name|Data Plotted|Description / Can Detect|Desirable | Not Desirable 
|:---|---|---|---|---
|**Residuals vs Fitted**| Fitted values (x-axis) <br/> Residuals (y-axis) |Used to check the assumption that relationships are linear.  Can detect if non-linear relationship was overlooked. | Horizontal line <br/> Randomness with dots (i.e., no distinct pattern)| Non-horizontal line <br/> Clumping of dots  
|**Normal Q-Q**|Theoretical Quantiles (x-axis) <br/> Standardized Residuals (y-axis) | Determines if the residuals are normally distributed.|Straight dashed line <br/> Residuals following the line|A curved dashed line <br/> Residuals not following the dashed line. 
|**Scale-Location <br/> (also known as Spread-Location)**|Fitted values (x-axis) <br/> SQRT of Standardized Residuals (y-axis)|Used to check the variance of the residuals. |Horizontal line <br/> Equally spread points (homoscedasticity) |Non horizontal line <br/> Patterns with the points (heteroscedasticity) 
|**Residual vs Leverage** | Leverage (x-axis) <br/> Standardized Residuals (y-axis) <br/> Cook's Distance (red dotted line)| Helps identify influencial cases (outliers), which may need to be removed from the analysis. | All cases are within Cook's Distance lines | Cases outside of Cook's Distance lines



```r
library(ggplot2)
library(gridExtra)
```

```r
#functions to improve code readability

CreateBasicHistogram <- function(v, bins, t){
    plot<- ggplot() + 
    theme_classic() + 
    geom_histogram(aes(v), bins = 30, fill='white', colour='black',size=0.5) + 
    labs(title = t)
    plot
}

CreateBasicPoints <- function(x_val, y_val, t){
    plot_points <- ggplot() + geom_point(aes(x_val, y_val)) +
    labs(title = t)
    plot_points
}

generateDataFrame <- function(numberOfRows){
    set.seed(123)
    x_norm <- rnorm(numberOfRows,5,7)
    df <- data.frame(matrix(ncol=6, nrow=0))
    df <- data.frame("id" = 0,
                     "norm" = x_norm,
                     "norm2" = x_norm^2,
                     "norm3" = x_norm^3,
                     "gamma" = dgamma(1:numberOfRows, shape = 1, rate = 0.01),
                     "binom" = rbinom(numberOfRows, 100, 0.2),
                     "error_norm" = rnorm(numberOfRows, 0, 2))
    df$id = 1:nrow(df)
    df
}

truth1 <- function(df1){
    beta0 = 1.1 
    beta1 = -8.2
    truth1 <- beta0 + beta1*df1$norm + df1$error_norm
    truth1
}

truth2 <- function(df1){
    beta0 = 1.1 
    beta1 = -8.2
    truth2 <- beta0 + beta1*df1$gamma + df1$error_norm
    truth2
}

truth3 <- function(df1){
    truth3 <- df1$norm2
    truth3
    
}
```

#### Generate the data and expore it

```r
df <- generateDataFrame(1000)
plot_x_norm <- CreateBasicHistogram(df$norm, 30, "norm" )
plot_x_norm2 <- CreateBasicHistogram(df$norm2, 30, "norm^2" )
plot_x_norm3 <- CreateBasicHistogram(df$norm3, 30, "norm^3" )
plot_x_gamma <- CreateBasicHistogram(df$gamma, 30, "gamma" )
plot_x_binom <- CreateBasicHistogram(df$binom, 30, "binom" )

grid.arrange(plot_x_norm, plot_x_norm2, plot_x_norm3, plot_x_gamma, plot_x_binom, nrow=3)
```

#### Generate the data, set truth1, and then visualize

```r
t1 = truth1(df)

plot_y <- CreateBasicHistogram(t1, 30, "When truth = 1.1 + -8.2norm + error " )
plot_norm_y <- CreateBasicPoints(df$norm, t1, "x (norm) compared to truth")
plot_gamma_y <- CreateBasicPoints(df$gamma, t1, "gamma compared to truth")
grid.arrange(plot_x_norm, plot_y, plot_norm_y, plot_gamma_y, nrow=2)
```

#### Create three models for truth1:
* Poor model that assumes a dependency on an irrelevant value from the dataframe
* Good model that assumes a dependency on only one of the relevant values from the dataframe
* Perfect model that assumes the two correct dependencies

```r
model_poor <- lm(t1 ~ df$gamma)
model_good <- lm(t1 ~ df$norm)
model_perfect <- lm(t1 ~ df$norm + df$error_norm)
```

#### Compare the coefficients, mean squared errors, and diagnostics from the three models

```r
compareModels <- function(poor, good, perfect, hasBetas){

    oldw <- getOption("warn")
    options(warn = -1)
    
    if (hasBetas){
        beta0 = 1.1 
        beta1 = -8.2
        print(paste("beta0 was set to ", beta0, "- Discovered with models poor, good, perfect:", summary(poor)$coefficients[1, 1], ",", summary(good)$coefficients[1, 1], ",", summary(perfect)$coefficients[1, 1]))
        print(paste("beta1 was set to", beta1, "- Discovered with models poor, good, perfect:", summary(poor)$coefficients[2, 1], ",", summary(good)$coefficients[2, 1], ",", summary(perfect)$coefficients[2, 1]))

    }
    
    print(paste("Mean Squared Error for models poor, good, perfect: ", mean(poor$residuals^2), ",", mean(good$residuals^2), ",", mean(perfect$residuals^2)))
    par(mfrow=c(4,3))
    plot(poor, 1, "Poor \n Residuals vs Fitted")
    plot(good, 1, "Good \n Residuals vs Fitted")
    plot(perfect, 1, "Perfect \n Residuals vs Fitted" )
    
    plot(poor, 2)
    plot(good, 2)
    plot(perfect, 2)
    
    plot(poor, 3)
    plot(good, 3)
    plot(perfect, 3)
    
    plot(poor, 5)
    plot(good, 5)
    plot(perfect, 5)
    
    options(warn = oldw)
}
compareModels(model_poor, model_good, model_perfect, TRUE)
```

#### Summary for truth1

* **Coefficients.** As expected, the Perfect model recovered the exact values of beta0 and beta1.
* **Mean Squared Error.**  As expected, the mean squared error is highest for the Poor model and lowest (essentially zero) for the Perfect model.
* **Residuals vs Fitted Interpretation.** The Poor model (that assumed an incorrect dependency on a gamma distribution) shows some clumping towards the right, which can indicate a problem.  However, the good and perfect models show more equally spread residuals along the red dotted line. 
* **Normal Q-Q Interpretation.**  All three models are following a straight line, but the Perfect model is staying close to zero
* **Scale-Location Interpretation.** The Poor model shows clumping on the right but the Good and Perfect models show a more equal spacing along the horizontal line.
* **Residual vs Leverage Intepretation.** None of the plots are showing extreme outliers outside of Cook's distance.

```r
t2 = truth2(df)
plot_x_gamma <- CreateBasicHistogram(df$gamma, 30, "x is gamma distribution" )
plot_truth2 <- CreateBasicHistogram(t2, 30, "Truth = 1.1 + -8.2gamma + error " )
plot_gamma_y <- CreateBasicPoints(df$gamma, t2, "gamma compared to truth")
grid.arrange(plot_x_gamma, plot_truth2, plot_gamma_y, nrow=2)
```

```r
model_poor2 <- lm(t2 ~ df$norm)
model_good2 <- lm(t2 ~ df$gamma)
model_perfect2 <- lm(t2 ~ df$gamma + df$error_norm)

compareModels(model_poor2, model_good2, model_perfect2, TRUE)
```

```r
t3 = truth3(df)
plot_x_norm2 <- CreateBasicHistogram(df$norm2, 30, "x is norm distribution squared" )
plot_truth3 <- CreateBasicHistogram(t3, 30, "When truth = x^2 " )
plot_x_y <- CreateBasicPoints(df$norm2, t3, "norm^2 compared to truth")
plot_x_y2 <- CreateBasicPoints(df$norm, t3, "norm compared to truth")
grid.arrange(plot_x_norm2, plot_truth3, plot_x_y, plot_x_y2, nrow=2)
```

```r
model_poor3 <- lm(t3 ~ df$norm3)
model_good3 <- lm(t3 ~ df$norm)
model_perfect3 <- lm(t3 ~ df$norm2)

compareModels(model_poor3, model_good3, model_perfect3, FALSE)
```

#### Summary for truth3 (x^2)

* **Mean Squared Error.**  As expected, the mean squared error is highest for the Poor model and lowest for the Perfect model.
* **Residuals vs Fitted Diagnostics.**  _Shows if the residuals have non-linear patterns._ The non-horizontal red lines for the Poor and Good models indicate a problem. 
* **Normal Q-Q.** _Determines if the residuals are normally distributed._  The Poor and Good models are not following a straight line which indicates that the residuals are not normally distributed.  
* **Scale-Location.** _Also known as Spread-Location, shows if there is equal variance among the residuals. A horizontal line with equally spread values is a good sign._  The Poor and Good models are showing a W-shape and 
* **Residual vs Leverage.** _Helps identify influencal cases, which may need to be removed from the analysis._  The Poor model is shwoing some values outside of Cook's distance, which means there are some influential cases.  The Good model is showing an influential case (164) in the upper right hand corner.  The Perfect model is showing no influential cases.
